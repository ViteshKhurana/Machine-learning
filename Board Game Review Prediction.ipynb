```python
# Remove any rows without user reviews.
df = df[df['users_rated'] > 0]
# Remove any rows with missing values.
df = df.dropna(axis = 0)
```

```python
# Get all the columns from the dataframe.
columns = df.columns.tolist()
# Filter the columns to remove ones we don't want.
columns = [c for c in columns if c not in ["bayes_average_rating","playingtime","average_rating", "type", "name", "id"]]

# Store the variable
X = df[columns]
y = df['average_rating']
```

```python
# Import a convenience function to split the sets.from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
# Import the linear regression model.
from sklearn.linear_model import LinearRegression
# Import the scikit-learn function to compute error.
from sklearn.metrics import mean_squared_error


# Initialize the model class.
l_model = LinearRegression().fit(X_train, y_train)

# Generate our predictions for the test set.
y_pred = l_model.predict(X_test)

# Compute error between our test predictions and the actual values.
mean_squared_error(y_pred, y_test)


```

```python
# Import the random forest model.
from sklearn.ensemble import RandomForestRegressor

# Initialize the model with some parameters.
r_model = RandomForestRegressor(n_estimators=100, min_samples_leaf=10, random_state=1).fit(X_train, y_train)

# Generate our predictions for the test set.
y_pred = r_model.predict(X_test
                    
                         
# Compute error between our test predictions and the actual values.                         
mean_squared_error(y_pred, y_test)
```

```python
#Import the Decision Tree model
from sklearn.tree import DecisionTreeRegressor

d_model = DecisionTreeRegressor(max_depth= 9).fit(X_train, y_train)


# Generate our predictions for the test set.
predd = d_model.predict(X_test)


# Compute error between our test predictions and the actual values.
mean_squared_error(predd, y_test)
```
